```python
import model_loader
import pipeline
from PIL import Image
from transformers import CLIPTokenizer
import torch

Device="cpu"

ALLOW_CUDA=False
ALLOW_MPS=False

if torch.cuda.is_available() and ALLOW_CUDA:
    Device="cuda"
elif ALLOW_MPS and hasattr(torch.backends, 'mps') and torch.backends.mps.is_available():
    print(f"Using {Device} device")

tokenizer=CLIPTokenizer("E:/Zju/CS/Diffusion model/data/vocab.json",merges_file="E:/Zju/CS/Diffusion model/data/merges.txt")
model_file="E:/Zju/CS/Diffusion model/data/v1-5-pruned-emaonly.ckpt"
models=model_loader.preload_models_from_standard_weights(model_file,Device)


prompt="A cat is running on a grassy field"
uncond_prompt=""
do_cfg=True
cfg_scale=7.5
input_image=None
image_path="E:/Zju/CS/Diffusion model/images/dog.png"
input_image=Image.open(image_path)
strength=0.8

sampler="ddpm"
num_inference_steps=50
seed=42
output_image=pipeline.generate(prompt=prompt,uncond_prompt=uncond_prompt,input_image=input_image,strength=strength,do_cfg=do_cfg,cfg_scale=cfg_scale,sampler_name=sampler,n_inference_steps=num_inference_steps,models=models,seed=seed,device=Device,tokenizer=tokenizer)

Image.fromarray(output_image)
```
